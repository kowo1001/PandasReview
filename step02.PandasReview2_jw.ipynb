```python
# 가격(Cost) 오름차순 정렬
data.sort_values(by='Cost', ascending=True)
```

```python
# 가격(Cost)에 따른 등급 책정 
# 75000 이상 A등급, 61500 이상 B등급, 50000 이상 C등급, 25000 이상 D등급 나머지 E등급

grade = []
for cost in data['Cost']:
    if int(cost.replace(cost[4:7],'00')[1:]) >= 75000:
        grade.append('A등급')
    elif int(cost.replace(cost[4:7],'00')[1:]) >= 61500:
        grade.append('B등급')
    elif int(cost.replace(cost[4:7],'00')[1:]) >= 50000:
        grade.append('C등급')
    elif int(cost.replace(cost[4:7],'00')[1:]) >= 25000:
        grade.append('D등급')
    else:
        grade.append('E등급')

data['Grade'] = grade
data
```

```python
# 구매일자인 1/20/14는 배송완료 처리 , 나머지는 주문대기중
import numpy as np

data['Order status'] = np.where(data['Purchase Date'] == '2/3/14', '배송완료', '주문대기중')
data
```
